# Information Visualization study

Hi!

Thank you for patricipating in my research. I am looking at different ways of animating dots from an initial to a target position. For instance, in the video below, we add a trace behind the moving dots to reinforce the trajectory.
You will be shown around 60 graphs and you will be asked to draw a few lines.

Let's take a look at an example of an animation. Just play the video below.

```python
video = io.open('animation.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```

There will always be 20 dots moving around.

13 dots are moving in a similar direction and they are travelling the same distance. We call these dots "the main trend". In the video below, the trajectory of the dots following the main trend is made visible (light green lines).

The other dots are deviating from this main trend, and either move to another direction or travel a different distance.

```python
video = io.open('animation_main.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```



## Some dots move in the opposite direction

In the video below, the dots moving in the opposite direction from the main trend are emphasized (trajectory shown with light red lines).

These dots always move parallel to the main trend and travel the same distance.



```python
video = io.open('animation_opp.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```



## Some dots travel a longer distance

In the video below, the dots travelling a longer distance than the overall distance of the main trend are emphasized (trajectory shown with light red lines).

```python
video = io.open('animation_long.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```



## Some dots move sideways

In the video below, the dots moving sideways from the main trend are emphasized (trajectory shown with light red lines).

```python
video = io.open('animation_side.mp4', 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```

You will be asked to answer on of 3 types of question:
        * Draw the trajectories of the dots moving in the opposite direction than the main trend
        * Draw the trajectories of the dots that travel a distance longer than the average motion of the main trend
        * Draw the trajectories of the dots moving sideways from the main trend

Each question has 1 to 3 right answers (i.e. 1 to 3 trajectories).

Try to answer the question as accurately as possible. There is no time limit to answer the questions. Skipping questions is allowed (just don't draw anything).

The video below demonstrates the possible actions in a trial.

In the experiment you will be able to play each animation twice. The first run will start automatically; and you can click on the animation window whenever you want to play the animation a second time.

After watching the animation for the second time, click again to see the question. Questions will appear on the bottom.

To answer the question, i.e. draw the trajectories of particular dots, left-click at the initial position of the relevant dot, and left-click again at the target position of the dot. This will draw a line, that will be considered as an answer. You can add up to 3 trajectories. Single dots don't count as an answer.

To edit your answer left-click and drag one extremity of the trajectory. You can also delete a trajectory with right click.

When you are done drawing your answer trajectories, press "Submit" button to see the next trial.

Let's practice this workflow on a few trials.

Switch to the next tab in the browser when you are ready.
